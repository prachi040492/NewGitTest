```python
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y=2*x
```

```python

y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='g', linestyle=":",linewidth=2)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x = np.arange(1,11)
```

```python
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='g',linestyle=':')
plt.plot(x,y2,color='r',linestyle='-.')
plt.xlabel('Multiple plots')
plt.ylabel('Red & Green')
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color='g',linestyle=':')
plt.subplot(2,1,1)
plt.plot(x,y2,color='r',linestyle='-.')
plt.show()
```

```python
#Bar Plot
student= {'Bob':67, 'Matt':34, 'Sam':34}
names=list(student.keys())
values=list(student.values())
# plt.bar(names,values)
plt.barh(names,values,color='g')
plt.title('Bar Plot')
plt.xlabel('names')
plt.ylabel('marks')
plt.grid(True)
plt.show()
```
