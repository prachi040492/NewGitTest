```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.describe()
```

```python
df1= df[['Trans Fat','Carbohydrates (% Daily Value)']]
df2 = df[['Calories','Calories from Fat','Total Fat','Total Fat (% Daily Value)','Saturated Fat','Saturated Fat (% Daily Value)']]
df3 = df[['Cholesterol','Cholesterol (% Daily Value)','Sodium','Carbohydrates']]
df4 = df[['Dietary Fiber','Dietary Fiber (% Daily Value)','Sugars','Protein']]
df5 = df[['Vitamin A (% Daily Value)','Vitamin C (% Daily Value)','Calcium (% Daily Value)','Iron (% Daily Value)']]
          
```

```python
sns.boxplot(data=df1,orient='h',palette='Set2') #Columns having outliers : Trans Fat,Carbohydrates (% Daily Value)
```

```python
sns.boxplot(data=df2,orient='h',palette='Set2') #Columns having outliers : Calories,Calories from Fat,Total Fat,Total Fat (% Daily Value)
```

```python
plt.figure(figsize=(12,10))
sns.boxplot(data=df3,orient='h',palette='Set2')
plt.xticks(np.arange(0,1800,100))
plt.xticks(rotation=90)
plt.show()
#Columns having outliers : Cholesterol,Cholesterol (% Daily Value),Sodium,Carbohydrates
```

```python
sns.boxplot(data=df4,orient='h',palette='Set2')
#Columns having outliers : Dietary Fiber (% Daily Value),Sugars,Protein
```

```python
sns.boxplot(data=df5,orient='h',palette='Set2')
#Columns having outliers : Vitamin A (% Daily Value),Vitamin C (% Daily Value),Calcium (% Daily Value),Iron (% Daily Value)
#INFERENCE: All columns have outliers except 3 : Dietary Fiber,Saturated Fat,Saturated Fat (% Daily Value)
```
